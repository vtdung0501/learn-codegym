```python
import random as rd
while True:
    current_number = 0
    current_player = ""
    num = rd.randint(0,1)
    if num == 0:
        current_player = 'human'
    else:
        current_player = 'computer'
    while current_number <= 21:
        print(current_number)
        if current_player == "human":
            print("Người chơi")
            play_choice = ""
            play_choice = input("Hãy nhập một số: ")
            while play_choice in ['1','2','3']:
                break
            current_number+=int(play_choice)
            if current_number >=21:
                print(current_number)
                print("You are lose")
                break
            else:
                current_player = 'computer'
        elif current_player == 'computer':
            print("Máy chơi")
            computer_choice = rd.randint(1,3)
            print(computer_choice)
            current_number+=computer_choice
            if current_number >=21:
                print(current_number)
                print("You are win")
                break
            else:
                current_player = 'human'
    print("Bạn có muốn chơi lại không")
    play_again = input('Chọn y or y')
    if play_again == 'y':
        continue
    else:
        break
```
